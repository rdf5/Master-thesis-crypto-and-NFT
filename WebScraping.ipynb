# Web Scraping NFT data 

Data collected from <a href="https://www.cryptoslam.io/">CryptoSlam</a>

***

```python
# Modules needed
import requests  # download a web page
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# Collection: BAYC
page = requests.get("https://web-api.cryptoslam.io/web-api/v1/sales/bored-ape-yacht-club/summary?groupingtype=daily")
raw_data=page.json()
```

```python
df=pd.json_normalize(raw_data['salesSummaryRecords'])
```

```python
df 
```

```python
fig = px.line(df, x='date', y='salesUSD', hover_data=['salesToken','totalTransactions','avgPriceUSD'], 
              labels={'date':'Date','salesUSD':'Sales (USD)'}, title = "BAYC sales volume data", markers=True)
fig.show()
```
